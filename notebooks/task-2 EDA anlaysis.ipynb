## Clean Our Data Using The IQR (Interquartile Range) Method

Code for Outlier Detection and Removal using IQR

```python
import pandas as pd

# Function to remove outliers using IQR
def remove_outliers_iqr(df, column, iqr_factor=1.5):
    
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile range
    
    # Define bounds for outliers
    lower_bound = Q1 - iqr_factor * IQR
    upper_bound = Q3 + iqr_factor * IQR
    
    # Filter out the outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return filtered_df

# List of numerical columns to clean (update this with your numerical columns)
numerical_columns = ['Amount', 'Value']  # Replace with your actual numerical columns

# Apply the IQR outlier removal to each numerical column cumulatively
df_cleaned = df.copy()
for col in numerical_columns:
    df_cleaned = remove_outliers_iqr(df_cleaned, col)

# Print the shape of the cleaned dataset to see how many rows were removed
print("Shape of the original dataset:", df.shape)
print("Shape of the cleaned dataset:", df_cleaned.shape)



# Save the cleaned dataset to a CSV file
df_cleaned.to_csv(r'C:\Users\MMM\Documents\10 Academy File\KAIM-Week-6\data\cleaned_dataset.csv', index=False)

# Show the first few rows of the cleaned dataset
df_cleaned.head()
```
